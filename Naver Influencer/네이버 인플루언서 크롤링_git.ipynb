```python
#블로그 주소 추출
blogList = []
num = 1
print(len(totaldf))
for links in totaldf.itertuples():
    data = requests.get('https://in.naver.com/' + links.ID)
    pageString = data.content
    soup = BeautifulSoup(pageString, 'html.parser')
    blogaddr = soup.findAll('a')
    list2 = []
    try:
        for addr in blogaddr:
            if 'blog' in addr['href'] and '=' not in addr['href'] and 'influencer_search' not in addr['href'] and 'blogID' not in addr['href']:
                if len(addr['href'].split('/')) == 4:
                    list2.append(addr['href'])
            else:
                pass
    except KeyError:
        pass
    try:
        blog = list2[0]
        blogList.append(blog)
    except (IndexError, KeyError):
        blogList.append('정보없음')
        blog = '정보없음'
    print(num, '/', len(totaldf), blog, links.ID)
    num += 1
    
totaldf['블로그URL'] = blogList
totaldf['naverID'] = totaldf['블로그URL'].str.split('/').str[-1]
totaldf['이메일'] = totaldf['naverID'] + '@naver.com'
totaldf['이메일'][totaldf['이메일'].str.contains('정보없음')] = '정보없음'
totaldf
```
